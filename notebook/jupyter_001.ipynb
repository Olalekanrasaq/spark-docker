```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .master("spark://spark-master:7077") \
    .appName("DockerJupyterSpark") \
    .getOrCreate()
```

```python
df = spark.read.csv("/opt/spark-data/tips.csv", header=True, inferSchema=True)
```

```python
df.show()
```

```python
selected_cols = df.select('total_bill', 'sex', 'day')
selected_cols.show()
```

```python
df.filter(df.sex == 'Male').show()
```

```python
df.groupBy('sex').agg({'total_bill': 'sum', 'tip': 'sum'}).show()
```

```python
df.withColumn('percent_tip', (df.tip / df.total_bill) * 100).show()
```
